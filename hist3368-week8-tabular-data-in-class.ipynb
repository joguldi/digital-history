```python
import string
congress['speech'] = congress['speech'].str.replace('[{}]'.format(string.punctuation), '')
```

```python
import nltk
from nltk.corpus import stopwords

nltk.download('stopwords')
stop = stopwords.words('english')
stopwords_regex = r'\b(?:{})\b'.format('|'.join(stop))
stopwords_regex
congress['stopworded'] = congress['speech'].str.replace(stopwords_regex, '')
congress.head()
```
